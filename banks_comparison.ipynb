## Loans to the US government

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker  # Import ticker for formatting

# Data for loans outstanding
loans_outstanding = {
    'L1,1790': 2000000,
    'L2,1792': 400000,
    'L3,1794': 1000000,
    'L4': 800000,
    'L5': 1000000,
    'L6 1794': 2000000,
    'L7': 800000,
    'L8,1795': 1000000,
    'L9,1799': 200000
}

loan_number = list(loans_outstanding.keys())
amount_of_loan = list(loans_outstanding.values())

# Create the figure and axes for the bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size

# Create the bar graph
plt.bar(loan_number, amount_of_loan, color='blue', width=0.5)
plt.xlabel('Loan Number')
plt.ylabel('Amount of Loan in Dollars ($)')
plt.title('1st US Bank - Loans to the US Government')

# Format y-axis to display as currency with commas and dollar sign
ax = plt.gca()  # Get current axis
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))  # Format as $ and commas

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()  # Display the bar graph
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Load the dataset
df = pd.read_csv("/Users/aleskamedrano/Desktop/GitHub/HistoricalEconomics/code/personal/aleska/Fall 2024/bank_table_data.csv", thousands=",")

# Convert 'Date' to datetime format
date = pd.to_datetime(df["Date"], infer_datetime_format=True)

# Fill NaN values with 0
df_filled = df.fillna(0)

# Extract the relevant columns
bank_united_states = df_filled["Bank of the United States"]
bank_boston_branch = df_filled["Boston branch"]
bank_new_york_branch = df_filled["New York"]
bank_baltimore_branch = df_filled["Baltimore"]
bank_charleston_branch = df_filled["Charleston"]
bank_massachusetts = df_filled["Bank of Massachusetts"]
bank_new_york = df_filled["Bank of New York"]
bank_north_america = df_filled["Bank of North America"]
bank_providence = df_filled["Bank of Providence"]
bank_maryland = df_filled["Bank of Maryland"]

# Create the plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot each line separately with different colors
plt.plot(date, bank_united_states, label="Bank of the United States", color='blue')
plt.plot(date, bank_boston_branch, label="Boston branch", color='orange')
plt.plot(date, bank_new_york_branch, label="New York Branch", color='green')
plt.plot(date, bank_baltimore_branch, label="Baltimore branch", color='red')
plt.plot(date, bank_charleston_branch, label="Charleston branch", color='purple')
plt.plot(date, bank_massachusetts, label="Bank of Massachusetts", color='brown')
plt.plot(date, bank_new_york, label="Bank of New York", color='pink')
plt.plot(date, bank_north_america, label="Bank of North America", color='cyan')
plt.plot(date, bank_providence, label="Bank of Providence", color='gray')
plt.plot(date, bank_maryland, label="Bank of Maryland", color='magenta')

# Add title and labels
plt.title("Statement of cash in the Treasury, showing monthly balance during (1792-1793)")
plt.xlabel("Month")
plt.ylabel("Monthly Balance in dollars ($)")

# Format y-axis as currency
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))

# Add a legend with smaller font size
plt.legend(loc="upper right", fontsize='small')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Load the dataset
df = pd.read_csv("/Users/aleskamedrano/Desktop/GitHub/HistoricalEconomics/code/personal/aleska/Fall 2024/bank_table_data.csv", thousands=",")

# Convert 'Date' to datetime format
date = pd.to_datetime(df["Date"], infer_datetime_format=True)

# Fill NaN values with 0
df_filled = df.fillna(0)

# Extract the relevant columns for the first plot
total = df_filled["Total"]
sum_bills = df_filled["Sums in bills as returned by Bank of the United States"]
specie = df_filled["Specie"]

# Extract the relevant columns for the second plot
bank_united_states = df_filled["Bank of the United States"]
bank_boston_branch = df_filled["Boston branch"]
bank_new_york_branch = df_filled["New York"]
bank_baltimore_branch = df_filled["Baltimore"]
bank_charleston_branch = df_filled["Charleston"]
bank_massachusetts = df_filled["Bank of Massachusetts"]
bank_new_york = df_filled["Bank of New York"]
bank_north_america = df_filled["Bank of North America"]
bank_providence = df_filled["Bank of Providence"]
bank_maryland = df_filled["Bank of Maryland"]

# Create the figure and subplots
plt.figure(figsize=(14, 6))  # Adjust the overall figure size

# First subplot: Total, Sums in Bills, and Specie
plt.subplot(1, 2, 1)
plt.plot(date, total, label="Total", color='blue')
plt.plot(date, sum_bills, label="Sums in bills", color='orange')
plt.plot(date, specie, label="Specie", color='green')
plt.title("Total, Sums in Bills, and Specie (1792-1793)")
plt.xlabel("Month")
plt.ylabel("Monthly Balance in dollars ($)")
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))  # Format y-axis
plt.legend(loc="lower left")

# Second subplot: Statement of cash by branch
plt.subplot(1, 2, 2)
plt.plot(date, bank_united_states, label="Bank of the United States", color='blue')
plt.plot(date, bank_boston_branch, label="Boston branch", color='orange')
plt.plot(date, bank_new_york_branch, label="New York Branch", color='green')
plt.plot(date, bank_baltimore_branch, label="Baltimore branch", color='red')
plt.plot(date, bank_charleston_branch, label="Charleston branch", color='purple')
plt.plot(date, bank_massachusetts, label="Bank of Massachusetts", color='brown')
plt.plot(date, bank_new_york, label="Bank of New York", color='pink')
plt.plot(date, bank_north_america, label="Bank of North America", color='cyan')
plt.plot(date, bank_providence, label="Bank of Providence", color='gray')
plt.plot(date, bank_maryland, label="Bank of Maryland", color='magenta')
plt.title("Statement of Cash by Branch (1792-1793)")
plt.xlabel("Month")
plt.ylabel("Monthly Balance in dollars ($)")
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))  # Format y-axis
plt.legend(loc="upper right", fontsize='small')

# Adjust the layout and display the plots side by side
plt.tight_layout()
plt.show()
```
